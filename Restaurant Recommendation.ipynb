```python
def restaurant_details_by_name(restaurant_name, data):
    restaurants= data[data['Restaurant Name'] == restaurant_name]
    if restaurants.empty:
        return "No Data Available."
    else:
        for index, row in restaurants.iterrows():
            print("Restaurant ID:", row['Restaurant ID'])
            print("Restaurant Name:", row['Restaurant Name'])
            print("City:", row['City'])
            print("Address:", row['Address'])
            print("Cuisines:", row['Cuisines'])
            print("Price range:", row['Price range'])
            print("Aggregate rating:", row['Aggregate rating'])
            print("Has Online delivery:", row['Has Online delivery'])
            print("\n")
    

userdata=input("Enter Restaurent Name: ")
restaurant_details_by_name(userdata, data)
```

```python
def restaurant_details_by_city(city, data):
    restaurants= data[data['City'] == city]
    if restaurants.empty:
        return "No data available."
    else:
        for index, row in restaurants.iterrows():
            print("Restaurant ID:", row['Restaurant ID'])
            print("Restaurant Name:", row['Restaurant Name'])
            print("City:", row['City'])
            print("Address:", row['Address'])
            print("Cuisines:", row['Cuisines'])
            print("Price range:", row['Price range'])
            print("Aggregate rating:", row['Aggregate rating'])
            print("Has Online delivery:", row['Has Online delivery'])
            print("\n")
    

user_data=input("Enter City: ")
restaurant_details_by_city(user_data, data)
```

```python
def restaurant_details_by_price(price, data):
    restaurants = data[data['Price range'] == price]
    if restaurants.empty:
        return "No data available."
    else:
        num_restaurants = min(10, len(restaurants))
        print(f"Total number of restaurants with price range {price}: {len(restaurants)}")
        print(f"Displaying a sample of {num_restaurants} restaurants:\n")
        for index, row in restaurants.head(10).iterrows():
            print("Restaurant ID:", row['Restaurant ID'])
            print("Restaurant Name:", row['Restaurant Name'])
            print("City:", row['City'])
            print("Address:", row['Address'])
            print("Cuisines:", row['Cuisines'])
            print("Price range:", row['Price range'])
            print("Aggregate rating:", row['Aggregate rating'])
            print("Has Online delivery:", row['Has Online delivery'])
            print("\n")
    

user_data=int(input("Enter price(1-4): "))
restaurant_details_by_price(user_data, data)
```

```python
df=data
df['sep_cuisines'] = df['Cuisines'].str.split(', ')
df= df.explode('sep_cuisines').reset_index(drop=True)
df
```

```python
def restaurant_details_by_cuisines(cuisines, data):
    restaurants = data[data['sep_cuisines'] == cuisines]
    if restaurants.empty:
        return "No data available."
    else:
        num_restaurants = min(10, len(restaurants))
        print(f"Total number of restaurants with cuisine {cuisines}: {len(restaurants)}")
        print(f"Displaying a sample of {num_restaurants} restaurants:\n")
        ids=set()
        for index, row in restaurants.head(10).iterrows():
            if(row['Restaurant ID'] not in ids):
                print("Restaurant ID:", row['Restaurant ID'])
                print("Restaurant Name:", row['Restaurant Name'])
                print("City:", row['City'])
                print("Address:", row['Address'])
                print("Cuisines:", row['Cuisines'])
                print("Price range:", row['Price range'])
                print("Aggregate rating:", row['Aggregate rating'])
                print("Has Online delivery:", row['Has Online delivery'])
                print("\n")
                ids.add(row['Restaurant ID'])
    

user_data=input("Enter cuisines you prefer: ")
restaurant_details_by_cuisines(user_data, df)
```

```python
def restaurant_details_by_online_delivery(online_delivery, data):
    delivery_status = "Yes" if online_delivery == 1 else "No"
    restaurants = data[data['Has Online delivery'] == delivery_status]
    if restaurants.empty:
        print(f"No restaurants offer online delivery: {delivery_status}")
    else:
        num_restaurants = min(10, len(restaurants))
        print(f"Total number of restaurants offering online delivery: {len(restaurants)}")
        print(f"Displaying a sample of {num_restaurants} restaurants that offer online delivery:\n")
        for index, row in restaurants.head(10).iterrows():
            print("Restaurant ID:", row['Restaurant ID'])
            print("Restaurant Name:", row['Restaurant Name'])
            print("City:", row['City'])
            print("Address:", row['Address'])
            print("Cuisines:", row['Cuisines'])
            print("Price range:", row['Price range'])
            print("Aggregate rating:", row['Aggregate rating'])
            print("Has Online delivery:", row['Has Online delivery'])
            print("\n")

user_data = int(input("Enter 1 for restaurants with online delivery, 0 otherwise: "))
restaurant_details_by_online_delivery(user_data, data)
```

```python
def call_function(day):
    match day:
        case 1:
            userdata=input("Enter Restaurent Name: ")
            restaurant_details_by_name(userdata, data)
        case 2:
            user_data=input("Enter City: ")
            restaurant_details_by_city(user_data, data)
        case 3:
            user_data=int(input("Enter price(1-4): "))
            restaurant_details_by_price(user_data, data)
        case 4:
            user_data=float(input("Enter minimum aggregate rating(0-4.9): "))
            restaurant_details_by_rating(user_data, data)
        case 5:
            user_data=input("Enter cuisines you prefer: ")
            restaurant_details_by_cuisines(user_data, df)
        case 6:
            user_data = int(input("Enter 1 for restaurants with online delivery, 0 otherwise: "))
            restaurant_details_by_online_delivery(user_data, data)
        case _:
            return "Invalid data"

print("Select the criteria for restaurant recommendations:")
print("option 1: By Name\noption 2: By City\noption 3: By Price\noption 4: By Rating\noption 5: By Cuisines\noption 6: By Online delivery availability")
call_function(int(input("")))
```
